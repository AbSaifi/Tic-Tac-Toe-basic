<p>    
1. Game is to be played between 2 players on a single PC<br>
2. Game is to be played with 'O' and 'X' only<br>
3. Game follows the below position format<br>
 
  &nbsp;&nbsp;&nbsp;&nbsp;1 | 2 | 3<br>
  &nbsp;&nbsp;&nbsp;&nbsp;----------<br>
  &nbsp;&nbsp;&nbsp;&nbsp;4 | 5 | 6<br>
  &nbsp;&nbsp;&nbsp;&nbsp;----------<br>
  &nbsp;&nbsp;&nbsp;&nbsp;7 | 8 | 9<br>
  
4.Player 1 gets to chose first whether he wants to start with 'X' or 'O'<br>
5.Players provide the position they want to chose on their turn<br>
6.Players alternate chosing positions until one of them wins or it is a draw<br>
7.Choice is given to play again or to exit<br>
8.All inputs to be provided in uppercase<br><br><br>

<b>To start game, execute function start_game()<br>
